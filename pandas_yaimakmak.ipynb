```python
import pandas as pd
```

```python
result_constituencies_PartyList = pd.read_excel("/content/drive/MyDrive/dataviz23data/ectreport66 (3).xlsx",sheet_name = 1)
result_constituencies_Candidate = pd.read_excel("/content/drive/MyDrive/dataviz23data/ectreport66 (3).xlsx",sheet_name = 2)
result_constituencies_status = pd.read_excel("/content/drive/MyDrive/dataviz23data/ectreport66 (3).xlsx",sheet_name = 3)
info_province = pd.read_excel("/content/drive/MyDrive/dataviz23data/ectreport66 (3).xlsx",sheet_name = 4)
info_party_overview = pd.read_excel("/content/drive/MyDrive/dataviz23data/ectreport66 (3).xlsx",sheet_name = 5)
info_constituency = pd.read_excel("/content/drive/MyDrive/dataviz23data/ectreport66 (3).xlsx",sheet_name = 6)
Candidate_PartyList = pd.read_excel("/content/drive/MyDrive/dataviz23data/ectreport66 (3).xlsx",sheet_name = 7)
Candidate_Constituency= pd.read_excel("/content/drive/MyDrive/dataviz23data/ectreport66 (3).xlsx",sheet_name = 8)
Candidate_PM = pd.read_excel("/content/drive/MyDrive/dataviz23data/ectreport66 (3).xlsx",sheet_name = 9)
```

```python
#หาผู้ชนะ สส เขต ในจังหวัด เเละคะเเนนเสียงที่ได้
winner = result_constituencies_Candidate[result_constituencies_Candidate["mp_app_rank"]==1]
```

```python
winner
```

```python
#เชื่อมจังหวัด เชื่อมคน เชื่อมพรรค
# จังหวัดใช้ con_id เเละ prov_id

result_constituencies_status
merge_pro = pd.merge(result_constituencies_status,winner,how="right",on=["cons_id"])
```

```python
merge_pro.drop(columns = ["percent_turn_out","counted_vote_stations","percent_count","pause_report"],inplace=True)
```

```python
merge_pro_final = pd.merge(merge_pro,info_province,how="left",on=["prov_id"])
```

```python
merge_pro_final.drop(columns=["abbre_thai","eng","province_id","total_vote_stations"],inplace=True)
```

```python
#เชื่อมคน mp_app_id
person_merge = pd.merge(merge_pro_final,Candidate_Constituency,how="left",on=["mp_app_id"])
```

```python
person_merge
```

```python
#เปลี่ยนชื่อ column เพราะชื่อไม่ตรงkey
info_party_overview.rename(columns={"id":"party_id"},inplace=True)
```

```python
#เชื่อมพรรค mp_app_party_id
party_merge = pd.merge(person_merge,info_party_overview,how="left",on=["party_id"])
```

```python
summary =party_merge[["mp_app_name","province","name","mp_app_vote"]].sort_values(by="mp_app_vote",ascending=False)
```

```python
kk_party = summary[(summary["name"]=="ก้าวไกล")]
```

```python
x = kk_party.value_counts("province").head(30)
```

```python
x
```

```python
count_kk_party_in_pro = pd.DataFrame({'total': x})
```

```python
count_kk_party_in_pro.reset_index(inplace=True)
```

```python
count_kk_party_in_pro
```

```python
#หาสส ขุนพล พรรค สัดส่วนเยอะสุด
percent = party_merge[["mp_app_name","mp_app_vote_percent","name","province"]].sort_values(by="mp_app_vote_percent",ascending=False)
```

```python
percent.reset_index(inplace=True)
percent
```

```python

percent
```

```python
percent[(percent["name"]=="เพื่อไทย")].head(10)
```

```python
lanslide = percent[(percent["mp_app_vote_percent"]>=50)].value_counts("name")
```

```python
lanslide_morethan_50 = pd.DataFrame({"total":lanslide})
```
